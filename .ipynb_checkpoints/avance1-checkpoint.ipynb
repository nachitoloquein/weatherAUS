## **Tratamiento de datos**

Crear nuevo data frame para mantentener los datos originales apartados

```python
data_frame_tratado = data_frame.copy()
```

Identificación de tipo de datos con sus columnas respectivas

```python
types = data_frame_tratado.columns.to_series().groupby(data_frame_tratado.dtypes).groups
type_obj = types[np.dtype('object')]
print('Existen',len(type_obj),'columnas de tipo objeto')
columnas = data_frame_tratado.columns
type_num = list(set(columnas) - set(type_obj))
print('Existen',len(type_num),'columnas de tipo numerico')
```

Se identifican la cantidad de columnas de cada tipo

Tratamiento nulls Numerico

```python
for columna in type_num:
  mean = data_frame_tratado[columna].mean()
  data_frame_tratado[columna] = data_frame_tratado[columna].fillna(mean) 
```

Para las columnas con datos de tipo cuantitativo (Float64) se reemplazan los datos nulls con el promedio de su respectiva columna. Esto con el fin de no perder datos valiosos en el DataFrame, pero puede llegar a contaminar la muestra.

Tratamiento null Object

```python
for columna in type_obj:
  mode = data_frame_tratado[columna].mode()[0]
  data_frame_tratado[columna] = data_frame_tratado[columna].fillna(mode) 
```

Para las columnas con datos de tipo categorico (Object) se reemplazan los datos nulls con la moda de su respectiva columna

Comprobación tratamiento null con exito

```python
for feature in data_frame.columns:
  print('Total de valores nulos de', feature, '=', data_frame_tratado[feature].isnull().sum())
```

Se puede apreciar la inexistencia de los datos null despues de realizar el tratamiento correspondiente.

Segunda Comprobación tratamiento de null con exito

```python
data_frame_tratado.isnull().any().any()
```

Segunda comprobación de existencia de nulls en el DataFrame, nos arroja False, lo que significa que de manera definitiva no existen datos nulls.

Codificación de datos categoricos

```python
#Opcional
df_codificado = data_frame_tratado

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_codificado['Location'] = le.fit_transform(data_frame_tratado['Location'])
df_codificado['WindGustDir'] = le.fit_transform(data_frame_tratado['WindGustDir'])
df_codificado['WindDir9am'] = le.fit_transform(data_frame_tratado['WindDir9am'])
df_codificado['WindDir3pm'] = le.fit_transform(data_frame_tratado['WindDir3pm'])
df_codificado['RainToday'] = le.fit_transform(data_frame_tratado['RainToday'])
df_codificado['RainTomorrow'] = le.fit_transform(data_frame_tratado['RainTomorrow'])

df_codificado.sample(10)
```

Se observa que los datos categoricos ahora se encuentran codificados para su uso estadistico.

## **Análisis de datos**

```python
plt.scatter(x = data_frame_tratado['Sunshine'], y = data_frame_tratado['MaxTemp'], edgecolors='black', color='#E5E8E8')
plt.title('Aumento de temperatura de acuerdo a las horas de Sol')
plt.xlabel('Horas de Sol')
plt.ylabel('Temperatura Maxima')
```

Se puede observar que existe una relación entra la temperatura y las hotas de sol, es decir mientras más horas de sol hayan durante el día, mayor será la temperatura de este.

```python
y, X = dmatrices('MaxTemp ~ Sunshine', data = data_frame_tratado, return_type = 'dataframe')
```

```python
modelo = sm.OLS(y, X)
resultado = modelo.fit()
print(resultado.summary())
```

***En este OLS podemos ver que la variable dependiente es la Temperatura maxima. En esta actividad se trabajó con el metodo Analisis Minimos cuadrados y contamos con la variable independiente (Df) horas de sol.***

***En este analisis podemos destacar diferentes puntos con sus respectivas interpretaciones:***

**R-squared (R^2):** *El 11% de la variabilidad del la temperatura maxima (Variable dependiente), es explicado por la variacion de las horas de sol (Variable independiente). Ademas hay un 89% de variabilidad no explicada.*

**Skew (coeficiente de asimetria):** *La distribución (0.200) es simetrica lo que significa que aprox. el 50% de los datos se encuentran a ambos lados de la media.*

**La Curtosis:** *Es de 2.9 por lo tanto la distribución es Leptocúrtica, es decir, existe alta concentración de datos en torno a la media.*

```python
plt.scatter(data_frame_tratado['Sunshine'], data_frame_tratado['MaxTemp'], edgecolors='black', color='#E5E8E8')
plt.plot(data_frame_tratado['Sunshine'], resultado.predict(), color="red", linewidth=3)
plt.title('Regresión Lineal Simple Temperatura respecto a Horas de Sol')
plt.xlabel('Horas de Sol')
plt.ylabel('Temperatura Maxima')
plt.show()
```

Se observa una leve relacion entre las variables, se puede deducir que ante mayor horas de sol tenga el dia, mayor temperatura Maxima habrá

```python
plt.stem(data_frame_tratado['Sunshine'], resultado.resid, use_line_collection=True);
plt.title("Gráfico de residuos")
plt.xlabel("Horas de Sol")
plt.ylabel("Residuo (error)")
plt.show()
```

Se puede obeservar que se obsevar valores normales en la gráfica de residuos de horas de sol.

```python
plt.scatter(x = data_frame_tratado['MinTemp'], y = data_frame_tratado['Pressure9am'], edgecolors='black', color='#E5E8E8')
plt.title('Grafico de Dispersión')
plt.xlabel('Temperatura Minima')
plt.ylabel('Presión a las 9 AM')
```

Utilizando las variables de temperatura y presión podemos llegar a la conclusión que se puede utilizar como indicador para predecir si lloverá el día de mañana o no, ya que a mayor presión y menor temperatura minima, mayor será la probabilidad que ocurran precipitaciones el día siguiente.

```python
y, X = dmatrices('Pressure9am ~ MinTemp', data = data_frame_tratado, return_type = 'dataframe')
```

```python
modelo = sm.OLS(y, X)
resultado = modelo.fit()
print(resultado.summary())
```

***En este OLS podemos ver que la variable dependiente es la Temperatura minima. En esta ocasión se trabajó con el metodo Analisis Minimos cuadrados y contamos con la variable independiente (Df) presión a las 9AM.***

***En este analisis podemos destacar diferentes puntos con sus respectivas interpretaciones:***

**R-squared (R^2):** *El 18% de la variabilidad en la temperatura minima (Variable dependiente), es explicado por la variacion de la presión a las 9AM (Variable independiente). Ademas hay un 82% de variabilidad no explicada.*

**Skew (coeficiente de asimetria):** *La distribución (-0.5) presenta asimetria negativa lo que significa que la mayoria de datos se concentras sobre la media.*

**La Curtosis:** *Es de 4.3 por lo tanto la distribución es Leptocúrtica, es decir, existe alta concentración de datos en torno a la media.*

Grafico de Regresión Lineal

```python
plt.scatter(data_frame_tratado['MinTemp'], data_frame_tratado['Pressure9am'], edgecolors='black', color='#E5E8E8')
plt.plot(data_frame_tratado['MinTemp'], resultado.predict(), color="red", linewidth=3)
plt.title('Gráfico de Presión Atmosfériva v/s La Temperatura Mínima')
plt.xlabel('Temperatura Minima')
plt.ylabel('Presión a las 9 AM')
plt.show()
```

Como se puede apreciar en el gráfico, existe una leve relación entre la presión atmosférica y la temperatura mínima de dicho día. Se puede asumir que los valores  que más se alejan de la linea de regresión líneal (Outlayers) es debido por causa de un fenómeno climático, por lo tanto podemos asumir que los outlayers no tendrían relevancia o no son significativos para predecir si lloverá el día de mañana o no.

```python
plt.stem(data_frame_tratado['MinTemp'], resultado.resid, use_line_collection=True);
plt.title("Gráfico de residuos")
plt.xlabel("Temperatura Minima")
plt.ylabel("Residuo (error)")
plt.show()
```

Podemos observar en el gráfico de residuos que el clima de Australia presenta una enorma variabilidad respecto a sus temperaturas, sin embargo podemos observar anomalias en donde hay temperaturas minimas mas altas de lo normal. Según lo anteriormencionado estos cambios en las temperaturas podrian ocurrir gracias a diversos fenomenos climaticos, como por ejemplo, incendios, huracanes u otros.

```python
rainfall_location_mean = data_frame_tratado.groupby('Location')['Rainfall'].mean()
rainfall_location_mean = rainfall_location_mean[:10].sort_values(ascending=False)
rainfall_location_mean.plot.bar(xlabel='Localidades', ylabel='Cantidad de lluvia en MM', title='TOP 10 localidades con más MM en promedio de lluvia', color='#1C2833')
```

Se puede observar con claridad la localidad **Cairns** la cual cuenta con un promedio de 5,7 MM de lluvia siendo esta la localidad con mayor MM de lluvias.

```python
rainfall_Dir_mean = data_frame_tratado.groupby('WindGustDir')['Rainfall'].mean()
rainfall_Dir_mean = rainfall_Dir_mean[:10].sort_values(ascending=False)
explode = [0.3,0.2,0.1,0,0,0,0,0,0,0]
rainfall_Dir_mean.plot.pie(title='Direcciones del viento con más MM en promedio de lluvia',autopct='%1.1f%%', shadow=True,figsize=(7,12),startangle=90, explode = explode)
```

Como se aprecia en el gráfico, la dirección hacia donde corre el viento que más cantidad de lluvia genera es el ***Sur***, en segundo lugar es el ***Sureste*** y en tercer lugar es el ***Estesureste***. Según lo anteriormente mencionado podemos concluir que la dirección del viento si influye en la predicicón de lluvia para el día de mañana, por lo tanto si el viento fluye hacia el sur o alguna variente de este, probablemente llueva el día de mañana.

```python
data_frame_date = data_frame_tratado.copy()
data_frame_date['Date'] = pd.to_datetime(data_frame_date['Date'],format='%Y-%m-%d') 
explode = [0,0,0,0.2,0,0,0,0,0,0,0]
rainfall_year = data_frame_date['Rainfall'].groupby(data_frame_date['Date'].dt.year).sum()
rainfall_year.plot.pie(x = rainfall_year, figsize=(7,12), autopct='%1.1f%%', explode = explode, shadow = True)
plt.title('Cantidad de MM de lluvia por año')
```

Según el gráfico de pastel o torta, podemos ver que la tendencia de la cantidad de lluvias por año es a la baja, es decir que la cantidad de lluvia que cae  va disminuyendo a traves de los año, ademas en el año 2017 el cual es el utlimo registrado presenta la mitad de las precitipaciones por metro cuadrado que en el año 2010. Si seguimos con esta tendencia bajista podemos concluir que los proximos años serán menos lluviosos y por lo tanto, tenderá a llover menos los días de invierno de los años futuros.
